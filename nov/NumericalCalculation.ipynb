```python
def f(x):
    return sin(x)

def diff_f(x, h):
    return (f(x+h)-f(x-h))/(2.0*h)
```

```python
# beta distribution
frozen_beta_true = beta.freeze(a=3, b=7, loc=2, scale=4)
data = frozen_beta_true.rvs(5000)
```

```python
fit = beta.fit(data)
print(fit)
```

```python
fit = beta.fit(data, floc=2, fscale=4)
print(fit)
```

```python
def f_ex3(x):
    # min y^2+2*z*y+z^2+10
    y = x[0]
    z = x[1]
    return y**2+2*z*y+z**2+10
```

```python
x0 = [0.3, 0.5]
```

```python
res3 = optimize.minimize(f_ex3, x0)
res3
```

```python
def f_ex4(x):
    y = x[0]
    z = x[1]
    return -(-z**2-y**2+z*y+z+y)
```

```python
x0 = [0.5, 0.5]
```

```python
res4 = optimize.minimize(f_ex4, x0=x0)
res4
```
