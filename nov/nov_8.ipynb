```python
# ベータ分布verを作成
def model_beta(Np, a, alpha, beta, m):
    gamma = (np.math.factorial(alpha-1) * np.math.factorial(beta-1)) / np.math.factorial(alpha+beta-1)
    func = lambda x: Np*(1 - scipy.integrate.quad(lambda y: (a**(alpha-1)*(1-alpha)**(beta-1))/gamma, 0, a)[0] * 
                         (1 - ((alpha+beta)/alpha)*(2*m*a)/(Np**(x-1)))**(Np-1))
    return func
```
