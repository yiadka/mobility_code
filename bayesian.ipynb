```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [12, 9]
```

```python
a = 56 + 1
b = 44 + 1
```

```python
x = np.linspace(0, 1, 100)
y = beta.pdf(x, a, b)
```

```python
print('Mean: ', a / (a + b))
print('Median: ', (a - 1) / (a + b - 2))
```

```python
plt.plot(x, y)
```

## 無情報事前分布を使う
$\beta$分布  
$\alpha = 1 + 1$

```python
a = 1
b = 1

x = np.linspace(0, 1, 100)
y = beta.pdf(x, a, b)
```

```python
print('Mean: ', a / (a + b))
```

```python
plt.plot(x, y)
```

上の表は一様分布である

```python
a = 57
b = 45

x = np.linspace(0, 1, 100)
y = beta.pdf(x, a, b)
```

```python
print('Mean: ', a / (a + b))
print('Mode: ', (a - 1) / (a + b - 2))
```

```python
plt.plot(x, y)
```

これは$Beta(\alpha=57,\beta=45)$のベータ分布であり、最頻値は0.56である
